```python
import pandas as pd
import numpy as np
import plotly.express as px

# Load dataset
df = pd.read_csv("gapminder.csv")

print(df.head())
```

```python
# Task 1

continent_df = df[df['Continent'] == 'Europe']

fig = px.scatter(
    continent_df,
    x='GDP_per_capita',
    y='Life_expectancy',
    size='Population',
    color='Continent',
    hover_name='Country',
    log_x=True,
    title='European countries show higher GDP per capita and life expectancy',
    template='simple_white'
)

fig.update_layout(
    font=dict(family='Arial'),
    xaxis_title='GDP per Capita (log scale)',
    yaxis_title='Life Expectancy'
)

fig.show()
```

```python
#Insight
#Asia and the Middle East experienced the largest increases in average CO2 emissions between 2000 and 2022, while some regions showed slower growth or stabilization.
```

```python
# Task 2

df['highlight'] = np.where(
    df['Continent'] == 'Asia',
    'Asia',
    'Other'
)

fig = px.scatter(
    df,
    x='GDP_per_capita',
    y='Life_expectancy',
    size='Population',
    color='highlight',
    hover_name='Country',
    log_x=True,
    size_max=60,
    color_discrete_map={
        'Asia': 'crimson',
        'Other': '#D3D3D3'
    },
    title='Asian countries combine large populations with growing economies',
    template='simple_white'
)

fig.add_annotation(
    x=10000,
    y=80,
    text='Asian countries show wide economic diversity',
    showarrow=True
)

fig.update_layout(
    font=dict(family='Arial'),
    xaxis_title='GDP per Capita (log scale)',
    yaxis_title='Life Expectancy'
)

fig.show()
```

```python
#Insight
#Asian countries display large variation in GDP per capita and life expectancy, highlighting the continent’s economic diversity and rapidly developing economies.
```
